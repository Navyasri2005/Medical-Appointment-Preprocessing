```python
import pandas as pd
```

```python
df = pd.read_csv("/content/KaggleV2-May-2016.csv",encoding='latin1')
```

```python
df = df.drop_duplicates()
df
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace('-', '_')
    .str.replace('hipertension', 'hypertension')
    .str.replace('handcap', 'handicap')
)
```

```python
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])


df['no_show'] = df['no_show'].str.strip().str.lower()


df['no_show'] = df['no_show'].map({'no': 0, 'yes': 1})


df = df.rename(columns={
    'sms_received': 'sms_received_flag',
    'scholarship': 'has_scholarship'
})
```

```python
df['patientid'] = df['patientid'].astype('int64').astype(str)


df['waiting_days'] = (df['appointmentday'] - df['scheduledday']).dt.days
```

```python
column_order = [
    'patientid', 'gender', 'age', 'neighbourhood',
    'scheduledday', 'appointmentday', 'waiting_days',
    'has_scholarship', 'hypertension', 'diabetes',
    'alcoholism', 'handicap', 'sms_received_flag',
    'no_show'
]
df
```

```python
df = df[column_order]
df.to_csv("Medical_Appointment_NoShows_Cleaned_Formatted.csv", index=False)
```
